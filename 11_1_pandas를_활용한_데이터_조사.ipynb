```python
import pandas as pd

#csv파일을 읽어와 데이터를 names 함수를 통해 속성별 키워드 지정
df = pd.read_csv('../content/pima-indians-diabetes.csv',names=["pregnant","plasma","pressure","thickness",
                                                               "insulin","BMI","pedigree","age","class"])
```

```python
#데이터의 첫 다섯 줄 출력
print(df.head(5))
```

```python
#데이터의 전반적인 정보 확인
print(df.info())
```

```python
#w정보별 특징 확인
print(df.describe())
```

```python
#일부의 데이터 확인
print(df[['pregnant', 'class']])
```

```python
#데이터 가공
#pregnant 속성을 기준으로 새그룹을 생성하고, 오름차순으로 정렬하고 각 pragnant에 해당하는 class 값의 평균을 나타남
print(df[['pregnant', 'class']].groupby(['pregnant'], as_index=False).mean().sort_values(by='pregnant', ascending=True))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,8)) #그래프의 크기 설정
sns.heatmap(df.corr(), linewidths=0.1, vmax=0.5, cmap=plt.cm.gist_heat, linecolor='white', annot=True) #heatmap은 각 속성의 상관관계를 나타내줌
plt.show()
```

```python
#plasma와 class 두 속성의 관게를 그래프로 확인
grid = sns.FacetGrid(df, col='class')
grid.map(plt.hist, 'plasma', bins=20) #bin은 막대의 수
plt.show()
```
